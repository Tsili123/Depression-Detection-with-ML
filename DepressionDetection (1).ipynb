```python
def tweet_normalization(text, remove_emojis=False, remove_emoticons=True, remove_links= False, remove_mentions = False, remove_hashtags= False):

    # remove links, hashtags, mentions
    if remove_links:
      text = re.sub(r'https?:/\/\S+', '', text)
    if remove_hashtags:
      text = re.sub(r"#(\w+)", '', text, flags=re.MULTILINE)
    if remove_mentions:
      text = re.sub(r"RT ", '', text, flags=re.MULTILINE)
      text = re.sub(r"@\w+", '', text, flags=re.MULTILINE)

    # convert text to lowercase
    text = text.lower()

    if not remove_emojis:
      #extract emojis from text and save them
      strv = ""
      for emoji in EMOJIS:
          if emoji in text:
              text = text.replace(emoji, "")
              strv += emoji

    if not remove_emoticons:
      strb = ""
      lista = []
      #approach 1: pattern for "generic smiley"
      eyes, noses, mouths = r":;8BX=", r"-~'^", r")(/\|*3DPO$"
      pattern1 = " [%s][%s]?[%s]" % tuple(map(re.escape, [eyes, noses, mouths]))

      lista=re.findall(pattern1, text)

      for var in lista:
        strb += var

    # remove non alphanumeric character
    text = re.sub(r'[^A-Za-z0-9]+', ' ', text)

    if not remove_emojis:
      #reset emojis to text
      text = text + strv

    if not remove_emoticons:
      #reset emoticons to text
      text = text + strb


    # apply blob techniques
    tweet_blob = TextBlob(text)

    # return normalized text
    blob_res = ' '.join(tweet_blob.words)

    # lematization techniques
    lemmatizer = WordNetLemmatizer()

    # stop words definition
    stop_words = nltk.corpus.stopwords.words('english')

    # combine all previous methods in order to clear the text
    result = ' '.join(
        lemmatizer.lemmatize(term)
        for term in text.split()
        if term not in set(stop_words)
    )

    return result
```

```python
def preprocess(input_df):
    # Remove rows with missing values in column col
    input_df.dropna(inplace=True)
    # Speed up code using numpy vectorization
    vfunc = np.vectorize(tweet_normalization)
    input_df["text_clean2"] = vfunc(input_df["post_text"])
    # return processed input_df
    return input_df
```

```python
import re
def getQuotes1(string):#mention
    return bool(re.findall("(@[\w_-]+)", string))

def getQuotes2(string):#hashtag
    return bool(re.findall("(#[\w_-]+)", string))

def getQuotes3(string):#links
    return bool(re.findall("(https?://)?([\w\-])+\.{1}([a-zA-Z]{2,63})([/\w-]*)*/?\??([^#\n\r]*)?#?([^\n\r]*)", string))

def getQuotes4(string):#emoticon
    return bool(re.findall(r"(?::|;|=)(?:-)?(?:\)|\(|D|P)", string))

def getQuotes5(string):#emoji
    return bool(re.findall(r'[\U00010000-\U0010ffff]', string))

df['Has_Mentions']=np.nan
df['Has_Mentions']=df['Has_Mentions'].astype('boolean')
df2 = df['post_text'].apply(lambda x: getQuotes1(x))
df['Has_Mentions'] = df2

df['Has_Hashtag']=np.nan
df['Has_Hashtag']=df['Has_Hashtag'].astype('boolean')
df2 = df['post_text'].apply(lambda x: getQuotes2(x))
df['Has_Hashtag'] = df2

df['Has_Links']=np.nan
df['Has_Links']=df['Has_Links'].astype('boolean')
df2 = df['post_text'].apply(lambda x: getQuotes3(x))
df['Has_Links'] = df2

df['Has_Emoticon']=np.nan
df['Has_Emoticon']=df['Has_Emoticon'].astype('boolean')
df2 = df['post_text'].apply(lambda x: getQuotes4(x))
df['Has_Emoticon'] = df2

df['Has_Emoji']=np.nan
df['Has_Emoji']=df['Has_Emoji'].astype('boolean')
df2 = df['post_text'].apply(lambda x: getQuotes5(x))
df['Has_Emoji'] = df2
```

```python
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder
from wordcloud import WordCloud

text = df['text_clean2'].str.cat(sep='\n')

nltk_tokens = nltk.word_tokenize(text)

finder = BigramCollocationFinder.from_words(nltk_tokens)
bigram_measures = BigramAssocMeasures()


scored = finder.score_ngrams(bigram_measures.raw_freq)
sorted_by_second = sorted(scored, key=lambda tup: tup[1], reverse=True)

print(scored[:10])

worddict = {}


listLen = len(sorted_by_second)


# Get the bigram and make a contiguous string for the dictionary key.
# Set the key to the scored value.
for i in range(listLen):
    worddict[' '.join(sorted_by_second[i][0])] = sorted_by_second[i][1]
# -----

# Set word cloud params and instantiate the word cloud.
# The height and width only affect the output image file.
WC_height = 2000
WC_width = 4000
WC_max_words = 40

wordCloud = WordCloud(max_words=WC_max_words, height=WC_height, width=WC_width)
wordCloud.generate_from_frequencies(worddict)
plt.title('Most frequently occurring bigrams for tweets')
plt.imshow(wordCloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import demoji

def convert_emojis_to_text(phrase):
    emoji_dict = demoji.findall(phrase)  # Finds all emojis and their descriptions
    for emoji, description in emoji_dict.items():
        # Replace each emoji with its description, convert spaces to underscores, and add a space at the end
        phrase = phrase.replace(emoji, description.lower().replace(" ", "_") + " ")
    return phrase
```

```python
from emot.emo_unicode import EMOTICONS_EMO
def convert_emoticons_to_text(phrase):
    # Iterate through each emoticon in the emoticons dictionary
    for emoticon in EMOTICONS_EMO:
        # Replace each emoticon in the phrase with its corresponding text representation, converted to lowercase
        phrase = phrase.replace(emoticon, EMOTICONS_EMO[emoticon].lower())
    return phrase
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

def generate_feature_vectors(vectorizer_type, training_data, testing_data):
    # Initialize the vectorizer variable to None
    text_vectorizer = None

    # Configure the vectorizer based on the specified type
    if vectorizer_type == "count":
        text_vectorizer = CountVectorizer(
            analyzer='word',
            ngram_range=(1, 2),  # Using both unigrams and bigrams
            token_pattern=r'[^\s]+')  # To include emojis and emoticons
    elif vectorizer_type == 'tf-idf':
        text_vectorizer = TfidfVectorizer(
            analyzer='word',
            ngram_range=(1, 2),  # Using both unigrams and bigrams
            token_pattern=r'[^\s]+')  # To include emojis and emoticons

    # Fit the vectorizer to the complete dataset to learn the vocabulary
    text_vectorizer.fit(sentences)

    # Transform the training and testing data into feature vectors
    training_vectors = text_vectorizer.transform(X_train)
    testing_vectors = text_vectorizer.transform(X_test)

    # Return the transformed feature vectors and the configured vectorizer
    return training_vectors, testing_vectors, text_vectorizer
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, roc_auc_score
from sklearn.naive_bayes import MultinomialNB
import numpy as np

def evaluate_naive_bayes_performance(training_features, training_targets, testing_features, testing_targets, smoothing_factor, use_prior):
    # Initialize the Multinomial Naive Bayes classifier with given parameters
    naive_bayes_classifier = MultinomialNB(alpha=smoothing_factor, fit_prior=use_prior)

    # Train the classifier on the training data
    naive_bayes_classifier.fit(training_features, training_targets)

    # Make predictions on the testing data
    predictions = naive_bayes_classifier.predict(testing_features)
    prediction_probabilities = naive_bayes_classifier.predict_proba(testing_features)

    # Convert probabilities to class labels for ROC AUC score calculation in binary classification scenarios
    if prediction_probabilities.shape[1] == 2:
        prediction_probabilities = np.array([np.argmax(prob) for prob in prediction_probabilities])

    # Calculate evaluation metrics
    accuracy_metric = accuracy_score(testing_targets, predictions)
    f1_metric = f1_score(testing_targets, predictions, average='weighted')
    auc_metric = roc_auc_score(testing_targets, prediction_probabilities, average='weighted', multi_class='ovr')

    # Return the evaluation metrics
    return accuracy_metric, f1_metric, auc_metric, naive_bayes_classifier
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, roc_auc_score
from sklearn import svm
import numpy as np
import time

def evaluate_svm_performance(training_features, training_labels, testing_features, testing_labels, c_value, kernel_type, gamma_value):
    # Initialize the Support Vector Machine classifier with the specified parameters
    svm_classifier = svm.SVC(C=c_value, kernel=kernel_type, gamma=gamma_value, probability=True)

    # Train the classifier on the training dataset
    svm_classifier.fit(training_features, training_labels)

    # Make predictions on the testing dataset
    predictions = svm_classifier.predict(testing_features)
    prediction_probabilities = svm_classifier.predict_proba(testing_features)

    # Adjust the prediction probabilities for binary classification to compute ROC AUC score
    if prediction_probabilities.shape[1] == 2:
        prediction_probabilities = np.array([np.argmax(prob) for prob in prediction_probabilities])

    # Calculate performance metrics
    accuracy_metric = accuracy_score(testing_labels, predictions)
    f1_metric = f1_score(testing_labels, predictions, average='weighted')
    auc_metric = roc_auc_score(testing_labels, prediction_probabilities, average='weighted', multi_class='ovr')

    # Return the computed metrics
    return accuracy_metric, f1_metric, auc_metric
```

```python
from sklearn.neighbors import KNeighborsClassifier

def evaluate_knn_performance(training_features, training_labels, testing_features, testing_labels, number_of_neighbors, weight_type):
    # Initialize the K-Nearest Neighbors classifier with specified parameters
    knn_classifier = KNeighborsClassifier(n_neighbors=number_of_neighbors, weights=weight_type)

    # Train the classifier using the training data
    knn_classifier.fit(training_features, training_labels)

    # Predict the labels for the testing data
    predictions = knn_classifier.predict(testing_features)
    prediction_probabilities = knn_classifier.predict_proba(testing_features)

    # For binary classification, convert probability predictions to class labels for ROC AUC score calculation
    if prediction_probabilities.shape[1] == 2:
        prediction_probabilities = np.array([np.argmax(prob) for prob in prediction_probabilities])

    # Compute performance metrics
    accuracy_metric = accuracy_score(testing_labels, predictions)
    f1_metric = f1_score(testing_labels, predictions, average='weighted')
    auc_metric = roc_auc_score(testing_labels, prediction_probabilities, average='weighted', multi_class='ovr')

    # Return the computed metrics
    return accuracy_metric, f1_metric, auc_metric
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold

# Function to Split Data into K-Folds for Experimentation
def split_data_into_k_folds(dataframe, target_labels, num_folds):
    folds_data = {}
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(target_labels)
    kfold_generator = KFold(n_splits=num_folds, shuffle=True, random_state=4)
    fold_index = 0
    for train_indices, test_indices in kfold_generator.split(dataframe.index):
        folds_data[fold_index] = {
            'train': dataframe.iloc[train_indices],
            'test': dataframe.iloc[test_indices],
            'y_train': encoded_labels[train_indices],
            'y_test': encoded_labels[test_indices]
        }
        fold_index += 1
    return folds_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from scipy.sparse import hstack

# Text Vectorization
text_vectorizer = TfidfVectorizer()
text_features = text_vectorizer.fit_transform(enhanced_df['text_clean2'])

# Normalization of Polarity Scores for MultinomialNB Compatibility
polarity_scaler = MinMaxScaler(feature_range=(0, 1))
normalized_polarity = polarity_scaler.fit_transform(enhanced_df[['Polarity']].values.reshape(-1, 1))

# Feature Combination: TF-IDF Text Features with Normalized Polarity
combined_features = hstack([text_features, normalized_polarity])

# Defining the Target Variable
target_labels = enhanced_df['label']

# Splitting Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(combined_features, target_labels, test_size=0.2, random_state=42)

# Training the Multinomial Naive Bayes Classifier
naive_bayes_classifier = MultinomialNB(alpha=0.12, fit_prior=False)
naive_bayes_classifier.fit(X_train, y_train)

# Model Prediction and Evaluation
model_predictions = naive_bayes_classifier.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, model_predictions))
print("Model Classification Report:\n", classification_report(y_test, model_predictions))
```
